# Make initial list of stars to feed to ccmap

```python
with open('reginitialcat.reg','w') as fileone:
    with open('initialUCAC5.txt','r') as file:
        next(file)
        for line in file:
            #print(line)
            text='physical; circle({},{},20)\n'.format(line.split()[2],line.split()[3])
            fileone.write(text)
```

```python
iraf.ccmap('initialUCAC5.txt', 'coord.db',results='initialccmapUCA5.txt',solutions='solinitial.txt',image='hubblef606wimg.fits',xcol=3,ycol=4,
           lngcol=1,latcol=2,lngunits='degrees',latunits='degrees',
           interactive='no',update='yes',insystem='icrs',projection='tan',fitgeometry='rxyscale',reject='3')
```

# From ds9 tsv to colum with ra and dec

# UCAC

```python
ra=[]
dec =[]
with open('UCAC5.tsv','r') as file:
    next(file)
    for line in file:
        ra.append(float(line.split()[0]))
        dec.append(float(line.split()[1]))
        
with open('UCAC5allradec.txt','w') as file:
    for i,j in zip(ra,dec):
        file.write("{} {}\n".format(i,j))
    
```

## Gaia

```python
ra=[]
dec =[]
with open('gaiaall.tsv','r') as file:
    next(file)
    for line in file:
        ra.append(float(line.split()[2]))
        dec.append(float(line.split()[4]))
        
with open('gaiaallradec.txt','w') as file:
    for i,j in zip(ra,dec):
        file.write("{} {}\n".format(i,j))
    
```

```python
catds9 = 'gaiaallradec.txt'
#catds9 = 'UCAC5allradec.txt'
#outputname = 'UCAC5matchedccfind'
outputnameccfind = 'gaiamatchedfind'


iraf.ccfind(catds9,outputnameccfind,'hubblef606wimg.fits',usewcs=True,lngcol=1,latcol=2,lngunits='degrees',latunits='degrees',
           insystem='icrs',refsystem='icrs',projection='tan',sbox=11,cbox=5)
```
